# Preprocessing Data Menggunakan Regex

## Menghapus Imbuhan

```python
# Mencari kata "dokternya" atau "Dokternya" terlebih dahulu
matches = df_alodokter['content'].str.contains(r'\b(dokternya|Dokternya)\b')
print("Baris yang mengandung 'dokternya' atau 'Dokternya':")
print(df_alodokter[matches])
```

```python
# Menghapus "-nya" dari "dokternya" atau "Dokternya"
df_alodokter['content'] = df_alodokter['content'].replace(r'\b(dokternya|Dokternya)\b', 'dokter', regex=True)
```

```python
print(df_alodokter[matches])
```

## Menggabungkan Kata

```python
# Mencari kata yang mengandung "terima" dan "kasih"
matches = df_alodokter['content'].str.contains(r'\bterima\s+kasih\b', case=False)
df_matches = df_alodokter[matches].reset_index(drop=True)
print("Baris yang mengandung 'terima' dan 'kasih':")
print(df_matches)
```

```python
# Menggabungkan "terima" dan "kasih"
df_alodokter['content'] = df_alodokter['content'].replace(r'\b[Tt]erima\s+[Kk]asih\b', 'terimakasih', regex=True)
```

```python
print(df_alodokter[matches])
```

## Melakukan Lowercasing

Tahap ini akan mengubah semua teks menjadi huruf kecil agar ketika melakukan analisis dapat lebih mudah karena bentuk kata telah homogen.

```python
print("Bentuk teks sebelum lowercasing:")
df_alodokter
```

```python
df_alodokter['content'] = df_alodokter['content'].str.lower()
```

```python
print("Bentuk teks setelah lowercasing:")
df_alodokter
```

## Menghapus Spasi Ekstra

Menghapus spasi berlebih antara kata-kata.

```python
df_alodokter['content'] = df_alodokter['content'].replace(r'\s+', ' ', regex=True).str.strip()
```

## Menghapus Tanda Baca

```python
# Mencari baris yang mengandung tanda baca
matches = df_alodokter['content'].str.contains(r'[^\w\s]', regex=True)
df_matches = df_alodokter[matches].reset_index(drop=True)

# Menampilkan hasil dalam tabel
print("Baris yang mengandung tanda baca:")
print(df_matches)
```

```python
# Menghapus tanda baca dari konten
df_alodokter['content'] = df_alodokter['content'].replace(r'[^\w\s]', '', regex=True)
```

```python
print("\nHasil akhir setelah penghapusan tanda baca:")
print(df_alodokter)
```

## Menghapus Stopwords

Menghapus kata-kata umum yang tidak memberikan makna (stop words).

```python
# Daftar stop words
stop_words = ['dan', 'atau', 'adalah', 'dg', 'saya', 'sy']
```

```python
# Mencari baris yang mengandung stop words
matches = df_alodokter['content'].str.contains(r'\b(?:' + '|'.join(stop_words) + r')\b', case=False, regex=True)
df_matches = df_alodokter[matches].reset_index(drop=True)

# Menampilkan hasil dalam tabel
print("Baris yang mengandung stop words:")
print(df_matches)
```

```python
# Menghapus stop words dari konten
df_alodokter['content'] = df_alodokter['content'].replace(r'\b(?:' + '|'.join(stop_words) + r')\b', '', regex=True)
```

## Menghapus Emoji

```python
# Menampilkan DataFrame sebelum penghapusan emoji
print("Sebelum penghapusan emoji:")
print(df_alodokter)
```

```python
# Menghapus emoji menggunakan regex
df_alodokter['content'] = df_alodokter['content'].replace(r'[^\w\s,.\'!"?]', '', regex=True)
```

```python
# Menampilkan DataFrame setelah penghapusan emoji
print("\nSetelah penghapusan emoji:")
print(df_alodokter)
```

## Normalisasi Kata (Stemming)

Mengubah variasi kata ke bentuk dasar.

```python
# Menangkap kata "pengobatan" atau "obat"
df_alodokter['content'] = df_alodokter['content'].replace(r'\b(pengobatan|obat)\b', 'obat', regex=True)
```

# Analisis Singkat

```python
# Mencari Kata Positif
def find_positive_words(df):
    pattern = r'\b(bagus|memuaskan|baik|bermanfaat|cepat|mudah|senang)\b'
    return df[df['content'].str.contains(pattern, case=False, regex=True)]

# Mencari Kata Negatif
def find_negative_words(df):
    pattern = r'\b(buruk|lambat|tidak puas|kesulitan|masalah)\b'
    return df[df['content'].str.contains(pattern, case=False, regex=True)]

# Mencari Kata Kunci "Alodokter"
def find_alodokter_mentions(df):
    pattern = r'\bAlodokter\b'
    return df[df['content'].str.contains(pattern, case=False, regex=True)]
```

```python
positif_reviews = find_positive_words(df_alodokter)
negatif_reviews = find_negative_words(df_alodokter)
alodokter_mentions = find_alodokter_mentions(df_alodokter)
```

```python
print("Ulasan Positif:", positif_reviews['content'].tolist())
print("Ulasan Negatif:", negatif_reviews['content'].tolist())
print("Sebutkan Alodokter:", alodokter_mentions['content'].tolist())
```

# Tokenizing

```python
import nltk
nltk.download('punkt')
```

```python
# Fungsi untuk melakukan tokenisasi pada setiap baris di kolom 'content'
df_alodokter['tokenized_content'] = df_alodokter['content'].apply(nltk.word_tokenize)
```

```python
# Menampilkan hasil tokenisasi
print(df_alodokter[['content', 'tokenized_content']])
```
